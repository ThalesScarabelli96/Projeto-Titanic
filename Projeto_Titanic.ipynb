Importando as bibliotecas necessárias

```python
import pandas as pd      # Para manipulação de dados (tabelas)
import seaborn as sns    # Para gráficos bonitos e avançados
import matplotlib.pyplot as plt  # Para suporte aos gráficos
```

Carregando os dados do Titanic

```python
# Carregando o dataset do Titanic
df = sns.load_dataset('titanic')

# Olhando as 5 primeiras linhas para entender a cara dos dados
df.head()
```

Entendendo o dataset

```python
# Verificando o número de linhas e colunas e vendo os tipos de colunas e se temos dados faltando (nulos)

print(f"O dataset possui {df.shape[0]} linhas e {df.shape[1]} colunas.")

df.info()
```

Limpeza Básica de Dados

```python
# Calculando a média de idade
media_idade = df['age'].mean()

# Preenchendo os valores nulos da coluna 'age' com a média
df['age'] = df['age'].fillna(media_idade)

# Confirmando que não há mais idades nulas
print(f"Dados nulos em Age: {df['age'].isnull().sum()}")
```

A Análise (Respondendo Perguntas com Dados)

* Pergunta 1:  Quem sobreviveu mais, homens ou mulheres?

```python
# Criando um gráfico de barras para ver a taxa de sobrevivência por sexo
# Nota: na coluna 'survived', 1 significa que sobreviveu e 0 que não.
sns.barplot(x='sex', y='survived', data=df, ci=None)
plt.title('Taxa de Sobrevivência por Sexo')
plt.ylabel('Proporção de Sobrevivência')
plt.xlabel('Sexo')
plt.show()
```

* Pergunta 2: A classe do bilhete influenciou na sobrevivência?

```python
# Criando um gráfico de contagem dividindo por classe e quem sobreviveu
sns.countplot(x='class', hue='survived', data=df)
plt.title('Sobreviventes vs Não Sobreviventes por Classe')
plt.ylabel('Quantidade de Passageiros')
plt.xlabel('Classe')
plt.legend(['Morreu', 'Sobreviveu'])
plt.show()
```

* Pergunta 3: Qual a idade média dos sobreviventes e dos que não sobreviveram?

```python
print("Idade média dos sobreviventes:", round(df[df['survived'] == 1]['age'].mean(), 2))
print("Idade média dos que não sobreviveram:", round(df[df['survived'] == 0]['age'].mean(), 2))
```

* Pergunta 4: Qual media do preço do bilhete dos sobreviventes e dos que não sobreviveram?

```python
print('Preço médio do bilhete dos sobreviventes:', round(df[df['survived'] == 1]['fare'].mean(), 2))
print('Preço médio do bilhete dos que não sobreviveram:', round(df[df['survived'] == 0]['fare'].mean(), 2))
```
